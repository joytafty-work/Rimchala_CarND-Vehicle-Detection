```python
from lesson_functions import convert_color, get_hog_features, color_hist, bin_spatial
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def dataset_metadata(zippath='data/vehicles.zip'): 
    with zipfile.ZipFile(zippath, 'r') as zf: 
        file_list = [
            imgfile for imgfile in zf.namelist() 
            if imgfile.endswith('.png')
            and not imgfile.startswith('__')
        ]
        
    metadata = pd.DataFrame(file_list, columns=['path'])
    metadata['set'] = metadata['path'].apply(lambda x: x.split('/')[0])
    metadata['source'] = metadata['path'].apply(lambda x: x.split('/')[1])
    metadata['image_name'] = metadata['path'].apply(lambda x: x.split('/')[-1])    
    
    return metadata[['set', 'source', 'image_name', 'path']]

def assemble_features_by_cspace(img, cspace, pix_per_cell, cell_per_block):
    color_converted = (
                cv2.cvtColor(img, eval('cv2.COLOR_RGB2{}'.format(cspace)))
                if cspace != 'RGB' else img
                )
    # Apply bin_spatial() to get spatial color features
    bin_features = bin_spatial(color_converted)
    # Apply color_hist() to get color histogram features
    color_hist_features = color_hist(color_converted)
    # Get HOG features
    hog_1 = get_hog_features(color_converted[:,:,0], orient, pix_per_cell, cell_per_block, vis=False, feature_vec=False)
    hog_2 = get_hog_features(color_converted[:,:,0], orient, pix_per_cell, cell_per_block, vis=False, feature_vec=False)
    hog_3 = get_hog_features(color_converted[:,:,0], orient, pix_per_cell, cell_per_block, vis=False, feature_vec=False)
    
    return np.hstack(
        [
            bin_features, color_hist_features, 
            hog_1, hog_2, hog_3
        ]
    )

def read_images(imgs, zippath=None):
    images = []
    if not zippath: 
        for img_file in imgs:
            images.append(mpimg.imread(img_file))
            
    else: 
        with zipfile.ZipFile(zippath, 'r') as zf: 
            for img_file in imgs: 
                images.append(plt.imread(BytesIO(zf.read(img_file))))
    
    return images

def extract_features(
    imgs, 
    cspace='YCrCb', 
    spatial_size=(32, 32),
    hist_bins=32, 
    hist_range=(0, 256), 
    zippath=None
):
    # Create a list to append feature vectors to
    features = []
    # Iterate through the list of images
    
    if not zippath: 
        for img_file in imgs:
            # Read in each one by one
            img = mpimg.imread(img_file)
            # apply color conversion if other than 'RGB'
            color_converted = (
                cv2.cvtColor(img, eval('cv2.COLOR_RGB2{}'.format(cspace)))
                if cspace != 'RGB' else img
                )
            # Apply bin_spatial() to get spatial color features
            bin_features = bin_spatial(color_converted)
            # Apply color_hist() to get color histogram features
            color_hist_features = color_hist(color_converted)
            # Append the new feature vector to the features list
            features.append(np.hstack([bin_features, color_hist_features]))
        # Return list of feature vectors
    else: 
        with zipfile.ZipFile(zippath, 'r') as zf: 
            for img_file in imgs: 
                img = plt.imread(BytesIO(zf.read(img_file)))
                color_converted = (
                    cv2.cvtColor(img, eval('cv2.COLOR_RGB2{}'.format(cspace)))
                    if cspace != 'RGB' else img
                )
                # Apply bin_spatial() to get spatial color features
                bin_features = bin_spatial(color_converted)
                # Apply color_hist() to get color histogram features
                color_hist_features = color_hist(color_converted)
                # Append the new feature vector to the features list
                features.append(np.hstack([bin_features, color_hist_features]))
    return features
```

```python
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from itertools import chain

def read_from_zipfile(zip_file_path, zip_file_metadata, image_indices):  
    images = []
    with zipfile.ZipFile(zip_file_path, 'r') as zf: 
        for iv in image_indices: 
            img_file = zip_file_metadata.loc[iv]['path']
            images.append(
                plt.imread(BytesIO(zf.read(img_file)))
            )
        return images

def sample_from_zipfile(zip_file_path, num_samples):
    metadata = dataset_metadata(zip_file_path)
    sample_indices = np.random.randint(low=0, high=len(metadata), size=num_samples)
    images = read_from_zipfile(
        zip_file_path=zip_file_path, 
        zip_file_metadata=metadata,
        image_indices=sample_indices
    )
    
    return images
    
vehicle_zip_file_path = 'data/vehicles.zip'
nonvehicle_zip_file_path = 'data/non-vehicles.zip'
num_samples = 16

# Sample some images to explore
sample_vehicle_images = sample_from_zipfile(
    zip_file_path=vehicle_zip_file_path,
    num_samples=num_samples
)

sample_nonvehicle_images = sample_from_zipfile(
    zip_file_path=nonvehicle_zip_file_path,
    num_samples=num_samples
)   
```

## Extract all features into DataFrames

```python
import time
from lesson_functions import convert_color, get_hog_features, color_hist, bin_spatial
from io import BytesIO
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
orient = 8
pix_per_cell = 8
cell_per_block = 2
spatial_size = (16, 16)
hist_bins = 32

vehicle_metada = dataset_metadata(zippath='data/vehicles.zip')
nonvehicle_metada = dataset_metadata(zippath='data/non-vehicles.zip')
cspace_list = ['RGB2YCrCb', 'RGB2HSV', 'RGB2HLS', 'RGB2LUV', 'RGB2YUV']

car_df = pd.DataFrame()
car_df['image'] = read_from_zipfile(
    zip_file_path='data/vehicles.zip',
    zip_file_metadata=vehicle_metada,
    image_indices=list(vehicle_metada.index)
)

for conv in cspace_list:
    start_time = time.time()
    # Convert color
    car_df['image_' + conv] = car_df['image'].apply(lambda x: convert_color(x))
    # Extract features by color
    # Spatial features
    car_df['bin_features_' + conv] = car_df['image_' + conv].apply(lambda x: bin_spatial(x))
    # Color historgram features
    car_df['color_hist_features_' + conv] = car_df['image_' + conv].apply(lambda x: color_hist(x))
    car_df['hog_1_' + conv] = car_df['image_' + conv].apply(lambda x: get_hog_features(x[:, :, 0], orient, pix_per_cell, cell_per_block, vis=True, feature_vec=True))
    car_df['hog_2_' + conv] = car_df['image_' + conv].apply(lambda x: get_hog_features(x[:, :, 1], orient, pix_per_cell, cell_per_block, vis=True, feature_vec=True))
    car_df['hog_3_' + conv] = car_df['image_' + conv].apply(lambda x: get_hog_features(x[:, :, 2], orient, pix_per_cell, cell_per_block, vis=True, feature_vec=True))
    elapsed_time = time.time() - start_time
    print('Elapsed time for channel {}: {} s'.format(
        conv, 
        time.strftime("%H:%M:%S", time.gmtime(elapsed_time))
    ))
```
